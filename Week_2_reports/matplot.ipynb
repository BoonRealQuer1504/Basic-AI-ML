```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('sampledata/california_housing_train.csv')
df
```

```python
plt.figure(dpi=75)
plt.hist(df['latitude'])
plt.title('Latitide distrinution')
plt.ylabel('Frequency')
plt.xlabel('Latitude')
```

```python
plt.figure(dpi=75)
plt.plot(df['median_income'][:10000])
```

```python
plt.figure(dpi=75)
plt.scatter(df['total_bedrooms'][:100], df['total_rooms'][:100])

plt.xlabel('Bedrooms')
plt.ylabel('Rooms')
plt.title('Rooms vs Bedrooms')
```

```python
plt.figure(dpi=75)
plt.scatter(df['total_bedrooms'][:100], df['total_rooms'][:100])
plt.scatter(df['total_bedrooms'][100:200], df['total_rooms'][100:200])
plt.xlabel('Bedrooms')
plt.ylabel('Rooms')
plt.title('Rooms vs Bedrooms')
plt.legend(['First hundred','Second hundred'])  ## phan chia nhom thanh cac object tren bieu do khac nhau
```

```python
bins_series = pd.cut(df['median_house_value'], bins=4).value_counts()
bins_series
```

```python
value=bins_series.to_list()
labels=[str(x) for x in bins_series.index.to_list()]
labels
```

```python
plt.pie(value,labels=labels,autopct='%1.2f%%')
plt.title('Distribution of median house values ')
```

```python
plt.bar(labels,height=value)
```

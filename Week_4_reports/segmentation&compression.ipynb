```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans

img=mpimg.imread('girl3.jpg')
plt.imshow(img)
imgplot=plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
X=img.reshape((img.shape[0]*img.shape[1],img.shape[2]))
##              row         height          rgb color 
## qua trinh flatten anh thanh mang 2D
## height*row=so pixel tong cong
## moi pixel sẽ có 3 màu (r,g,b)

```

```python
for K in [1,2,3,4,5,10,15,100]: # vong lap 5 lan voi gia tri dinh truoc 
    kmeans=KMeans(n_clusters=K).fit(X) # so cum =K, hàm fit phân K tâm cụm dựa trên dữ liệu có sẵn của ảnh
    label=kmeans.predict(X)  # đưa ra dự đoán các pixel thuộc cụm nào dựa trên tâm cụm ở trên

    img4=np.zeros_like(X) # lưu trữ ảnh sau khi phân cụm 
    for k in range(K):
        img4[label==k]=kmeans.cluster_centers_[k]
        ## cập nhật lại giá trị pixel thuộc cụm k trong img4, tất cả các pixel cùng 1 cụm có cùng màu 

    img5=img4.reshape((img.shape[0],img.shape[1],img.shape[2])) # chuyển về dạng ảnh ban đầu 
    plt.imshow(img5, interpolation='nearest')
    plt.axis('off')
    plt.show()
```

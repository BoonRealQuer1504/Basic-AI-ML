```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans

img=mpimg.imread('girl3.jpg')
plt.imshow(img)
imgplot=plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
X=img.reshape((img.shape[0]*img.shape[1],img.shape[2]))
##              row         height          rgb color 
## qua trinh flatten anh thanh mang 2D
## height*row=so pixel tong cong
## moi pixel s c 3 mu (r,g,b)

```

```python
for K in [1,2,3,4,5,10,15,100]: # vong lap 5 lan voi gia tri dinh truoc 
    kmeans=KMeans(n_clusters=K).fit(X) # so cum =K, hm fit phn K tm cm da trn d liu c sn ca nh
    label=kmeans.predict(X)  # a ra d on cc pixel thuc cm no da trn tm cm  trn

    img4=np.zeros_like(X) # lu tr nh sau khi phn cm 
    for k in range(K):
        img4[label==k]=kmeans.cluster_centers_[k]
        ## cp nht li gi tr pixel thuc cm k trong img4, tt c cc pixel cng 1 cm c cng mu 

    img5=img4.reshape((img.shape[0],img.shape[1],img.shape[2])) # chuyn v dng nh ban u 
    plt.imshow(img5, interpolation='nearest')
    plt.axis('off')
    plt.show()
```
